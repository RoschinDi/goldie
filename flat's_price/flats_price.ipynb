## Предобработка данных

### Переименовывание столбцов

Переименуем столбцы с нарушенным стилем и проверим

```python
data = data.rename(columns=
    {'cityCenters_nearest': 'city_centers_nearest',
    'parks_around3000': 'parks_around_3000',
    'ponds_around3000': 'ponds_around_3000'
    })
data.columns
```

### Обработка пропусков

Определим количество пропущенных значений в датафрейме

```python
data.isna().sum()
```

Пропуски обнаружились больше, чем у половины столбцов датафрейма. Необходимо проанализировать каждый столбец датафрейма, где присутствуют пропуски, на предмет возможности их обработки. Учитывая общую информацию о датафрейме (всего строк 23 699), можно заметить, что в некоторых столбцах пропуски занимают половину, а иногда и больше половины значений столбцов. Однако в некоторых столбцах таких, как `floors_total` и `locality_name`, количество пропусков невелико. С них и начнем.

Заполнить этажность домов и название населенного пункта в пропущенных значениях столбцов `floors_total` и `locality_name` не представляется возможным, т.к. возможно пользователь резмещая объявление попросту мог не указать этажность дома или название населенного пункта. Также данные пропуски могут быть системными сбоями при выгрузке данных. Удалим данные пропуски ввиду малого количества по данным столбцам.

```python
data.dropna(subset=['floors_total', 'locality_name'], inplace=True)
data.isna().sum()
```

Далее на очереди обработка пропусков в столбце `ceiling_height`. Пропуски скорее всего связаны с тем, что пользователи попросту не указывали высоту потолков при размещениии объявлений.

```python
#выведем числовое описание столбца
data['ceiling_height'].describe()
```

```python
#посчитаем процент пропусков от общего количества значений в столбце
f"{len(data[data['ceiling_height'].isna()]) / len(data['ceiling_height']):.2%}"
```

Так как количество пропусков в данном столбце составляет 38.55% от общего количества значений, то оставим их как есть, чтобы не искажать статистики.

Далее обработка пропусков в столбце `living_area`. Пропуски также связаны с тем, что пользователи попросту не указывали высоту потолков при размещениии объявлений.

```python
#выведем числовое описание столбца
data['living_area'].describe()
```

Напишем цикл, который при каждой итерации будет заполнять пропуски медианой, сгруппированной по количеству комнат

```python
for state in data['rooms'].unique():
        data.loc[(data['rooms'] == state) & (data['living_area'].isna()), 'living_area'] = \
        data.loc[data['rooms'] == state, 'living_area'].median()

#проверка, что пропусков не осталось
data['living_area'].isna().sum()
```

Проведем логическую проверку, что жилая площадь у нас не больше общей площади

```python
data.query('living_area > total_area')['living_area'].count()
```

По этим 29 объектам заполним значение жилой площади значением общей площади

```python
data.loc[data['living_area'] > data['total_area'],'living_area'] = data.loc[(data['living_area'] > data['total_area']),'total_area']

#проверка, что нет логических ошибок
data.query('living_area > total_area')['living_area'].count()
```

Обработка пропусков в столбце `is_apartment`. Столбец с булевыми значениями, значит пользователь при размещении объявления, считая, что его объект недвижимости не относится к апартаментам не заполнял данное поле. 

```python
#посмотрим на уникальные значения данного столбца
data['is_apartment'].value_counts()
```

Заполним значениями False, т.к. True только в 50 случаях.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)

#проверка, что пропусков не осталось
data['is_apartment'].isna().sum()
```

Обработка пропусков в столбце `kitchen_area`. Пропуски также связаны с тем, что пользователи не указывали площадь кухни при размещении объявлений.

```python
#выведем числовое описание столбца
data['kitchen_area'].describe()
```

Напишем цикл, который при каждой итерации будет заполнять пропуски медианой, сгруппированной по количеству комнат

```python
for state in data['rooms'].unique():
        data.loc[(data['rooms'] == state) & (data['kitchen_area'].isna()), 'kitchen_area'] = \
        data.loc[data['rooms'] == state, 'kitchen_area'].median()

#проверка, что пропусков не осталось
data['kitchen_area'].isna().sum()
```

Пропуски в столбце `balcony` заполним нулями, т.к. не указав никаких данных в данном столбце в большинстве случаев балкон (лоджия) у объекта недвижимости отсутствует

```python
data['balcony'] = data['balcony'].fillna(0)

#проверка, что пропусков не осталось
data['balcony'].isna().sum()
```

Т.к. пользователь при размещении объявления стремится указать любые преимущества реализуемого объекта недвижимости (а парки и пруды являются преимуществом), то пропуски в столбцах `parks_around_3000`, `ponds_around_3000` будем считать за отсутствие парков и прудов в радиусе 3 км от таких объектов недвижимости, поэтому вместо пропусков проставим  0.

```python
data['parks_around_3000'] = data['parks_around_3000'].fillna(0)
data['ponds_around_3000'] = data['ponds_around_3000'].fillna(0)

#проверка, что пропусков не осталось
data[['parks_around_3000', 'ponds_around_3000']].isna().sum()
```

Заполнить пропуски в остальных картографических данных не представляется возможным, поэтому пропуски в столбцах `airports_nearest`, `city_centers_nearest`, `ponds_nearest` и `parks_nearest`оставим без изменений.

Пропуски в столбце `days_exposition` будем считать, что не проставлены, потому что данные объявления еще не закрыты на момент выгрузки датасета, поэтому данные пропуски заполним количеством дней между датой размещения такого объявления в столбце `first_day_exposition` и самой поздней датой размещения в столбце `first_day_exposition` 

```python
#для начала переведем данные в столбце first_day_exposition в формат datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d %H:%M:%S')

#создадим переменную для самого позднего дня размещения для облегчения последующего кода
first_day_max = data['first_day_exposition'].max()

#заменим пропуски разницей между самым поздним первым днем размещения и первым днем размещения
data.loc[data['days_exposition'].isna(),
       'days_exposition'] = (first_day_max - data.loc[data['days_exposition'].isna(),
       'first_day_exposition']).dt.days

#проверка, что пропусков не осталось
data['days_exposition'].isna().sum()
```

### Изменение типов данных

Посмотрим еще раз на общую информацию о датафрейме

```python
data.info()
```

Переведем данные типа `float`  в тип `int` в тех столбцах, в которых по смыслу не может быть вещественных значений. Это столбцы: `floors_total`, `balcony`, `days_exposition`, `ponds_around_3000`, `parks_around_3000`. 

Также переведем из `float` в `int` значения столбца `last_price`, т.к. цену в объявлениях всегда указывают в рублях без копеек. 

```python
data_format = ['last_price', 'floors_total', 'balcony', 'parks_around_3000', 'ponds_around_3000',  'days_exposition']

data[data_format] = data[data_format].astype('int')

#проверим, что тип данных изменен в нужных столбцах
data.info()
```

### Устранение дубликатов

Проверим на наличие явных дубликатов.

```python
data.duplicated().sum()
```

Проверим столбец со строковыми значениями на наличие неявных дубликатов в столбце `locality_name`.

```python
data['locality_name'].unique()
```

Видно различающиеся по написанию, но одинаковые по смыслу названия населенных пунктов: `посёлок` и `поселок`. Приведем к одному стилю написания: заменим `посёлок` на `поселок`, но для начала посчитаем количество уникальных значений.

```python
len(data['locality_name'].unique())
```

```python
data['locality_name'] = data['locality_name'].str.replace('посёлок', 'поселок')

#проверяем, что действительно произошла замена
len(data['locality_name'].unique())
```

Анализируем дальше неявные дубликаты в столбце `locality_name`.

```python
data['locality_name'].unique()
```

Также заметно, что встречаются разные написания типов населенных пунктов, такие как: городской поселок и поселок городского типа, поселок и поселок станции или поселок при железнодорожной станции, коттеджный поселок. Приведем их к общему написанию. 

```python
data['locality_name'] = data['locality_name'].str.replace('городской поселок', 'поселок городского типа')
data['locality_name'] = data['locality_name'].str.replace('поселок станции', 'поселок')
data['locality_name'] = data['locality_name'].str.replace('поселок при железнодорожной станции', 'поселок')
data['locality_name'] = data['locality_name'].str.replace('коттеджный поселок', 'поселок')

#проверяем, что действительно произошла замена
len(data['locality_name'].unique())
```

## Посчитайте и добавьте в таблицу новые столбцы

Добавим к датафрейму столбец `price_for_1m`, который будет отображать цену за 1 кв. метр.

```python
#цену за 1 кв. метр найдем путем деления столбца цены на общую площадь
data['price_for_1m'] = data['last_price'] / data['total_area']

#изменим тип данных нового столбца на int
data['price_for_1m'] = data['price_for_1m'].astype('int')

#выведем 5 первых строк посмотреть
data.head()
```

Добавим к датафрейму столбцы `weekday_exposition`, `month_exposition`, `year_exposition`, которые будут отображать день публикации, месяц публикации и год публикации соответственно.

```python
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year

#выведем 5 первых строк посмотреть
data.head()
```

Добавим к датафрейму столбец `floor_type`, который будет отображать тип этажа в зависимости от того на каком этаже находится объект недвижимости: если в столбце floor стоит 1, то значит "первый", если в этом же столбце номер этажа равен этажности всего дома, значит "последний". Остальные этажи будут отнесены, как "другие".

```python
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'

data['floor_type'] = data.apply(floor_type, axis=1)

#выведем 5 первых строк посмотреть
data.head()
```

Добавим столбец `distance_center` и заполним его расстоянием в км от объекта недвижимости до центра, где это возможно

```python
data['distance_center'] = data['city_centers_nearest'] / 1000 

#выведем 5 первых строк посмотреть
data.head()
```
